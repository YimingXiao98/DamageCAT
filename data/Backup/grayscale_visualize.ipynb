```python
import matplotlib.pyplot as plt
import numpy as np
import PIL

# Assuming 'image' is your grayscale image as a 2D numpy array
# Example: image = np.array([[0, 1, 2], [3, 4, 0], [0, 0, 0]])
image = np.array(PIL.Image.open('/home/grads/y/yxiao/Personal Folder/Ida-Category/post_22.png').convert('L'))
# Create a color map
cmap = plt.cm.get_cmap('viridis', 5)  # 'viridis' is a perceptually uniform colormap

# Normalize the image to the range [0, 4]
norm_image = np.clip(image, 0, 4)

# Display the image with the color map
plt.imshow(norm_image, cmap=cmap, vmin=0, vmax=4)
plt.colorbar(ticks=[0, 1, 2, 3, 4], label='Pixel Value')
plt.title('Enhanced Grayscale Image')
plt.show()
```
