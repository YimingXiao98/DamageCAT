```python
from PIL import Image
import numpy as np

# Load the mask files
mask1 = Image.open('./post_101.png')
mask2 = Image.open('./hurricane-harvey_00000007_post_disaster.png')

# Print basic properties
print("Mask 1 - Size:", mask1.size, "Mode:", mask1.mode)
print("Mask 2 - Size:", mask2.size, "Mode:", mask2.mode)

# Convert images to numpy arrays for pixel comparison
mask1_array = np.array(mask1)
mask2_array = np.array(mask2)

# Check if the shapes of the arrays are the same
if mask1_array.shape != mask2_array.shape:
    print("The masks have different shapes.")
else:
    print("The masks have the same shape.")
    
np.unique(mask2_array)
```

```python
import matplotlib.pyplot as plt

# Define a colormap for the classes
cmap = plt.get_cmap('viridis', 5)  # 5 discrete colors

# Plot mask1
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(mask1_array, cmap=cmap, vmin=0, vmax=4)
plt.title('Mask 1')
plt.colorbar(ticks=[0, 1, 2, 3, 4])

# Plot mask2
plt.subplot(1, 2, 2)
plt.imshow(mask2_array, cmap=cmap, vmin=0, vmax=4)
plt.title('Mask 2')
plt.colorbar(ticks=[0, 1, 2, 3, 4])

plt.show()
```
